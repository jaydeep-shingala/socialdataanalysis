## Getting all the followers and their information from the given userid

```python
screen_name = name
userlist  = [name]

userids = [name]
for follower in api.followers(screen_name):
    if follower.screen_name not in userlist:
        userlist.append(follower.screen_name)
    if follower.screen_name not in userids:
        userids.append(follower.screen_name)

newlist = userlist[0:3]
for u in newlist:
    for follower in api.followers(u):
        if follower.screen_name not in userids:
            userids.append(follower.screen_name)
    
```
